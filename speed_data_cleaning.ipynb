## Import Necessary Libraries

```python
import pandas as pd
```

## Start With Cleaning

### Read the data

```python
dirty_data = pd.read_csv('dirty_data.csv')
dirty_data.head(20)
```

### Delete Unnamed Columns

```python
drop_columns_df = dirty_data.drop(columns = ['Unnamed: 9','Unnamed: 10'])
drop_columns_df
```

### First we will slice the data to groups
Slice by "first name: Person" row, and we do that by counter will be added to the dataframe
1. Drop nan values in Row type column

```python
drop_nan_df = drop_columns_df[drop_columns_df['Row Type'].notna()]
drop_nan_df
```

Create the counter 

```python
column_values = []
counter = 0

for i in drop_nan_df['Row Type']:
    if "first name" in i:
        counter +=1
    column_values.append(counter)
```

Assign the "Iteration" column and store the data in

```python
iter_cols = drop_nan_df
iter_cols['Iteration'] = column_values
iter_cols.head(20)
```

### Great, the data is grouped
Now, we will drop "Row Type" rows like index number 2 and 16

```python
drop_dirty_rows = iter_cols[iter_cols['Row Type'] != 'Row Type']
drop_dirty_rows
```

Let us get "first name: Person	last name: Human	date: end of time" rows and set it up to columns

```python
name_df = drop_dirty_rows[drop_dirty_rows['Row Type'].str.contains('first name')]
name_df.drop(columns = ['Speed1', 'Speed2', 'Electricity','Effort', 'Weight', 'Torque'],inplace = True)
name_df.rename(columns = {'Row Type':'First Name','Iter Number':'Last Name','Power1':'Date'},inplace = True)
name_df['First Name'] = name_df['First Name'].str.replace('first name: ','')
name_df['Last Name'] = name_df['Last Name'].str.replace('last name: ','')
name_df['Date'] = name_df['Date'].str.replace('date: ','')
name_df
```

Cool, now we have two dataframe [names DF, Speeds and others DF]

```python
no_name_df = drop_dirty_rows[~drop_dirty_rows['Row Type'].str.contains('first name')]
no_name_df
```

### Let's merge the data

```python
all_merged_df = pd.merge(left = no_name_df, right = name_df, how='inner', on='Iteration')
```

## WE DONE
### Here is the data after cleaning

```python
all_merged_df
```

### Convert the dataframe to CSV file

```python
all_merged_df.to_csv('clean_data.csv', index=False)
```
